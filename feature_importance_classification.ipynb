```python
# copy the dataset to df2
df2=dataset1
```

```python
# Convert the categorical data to numerical data using get_dummies
df2 = pd.get_dummies(df2, drop_first=True)
```

```python
# Get independent and dependent variables
indep_X=df2.drop(columns='classification_yes', axis=1)
dep_Y=df2['classification_yes']
```

```python
# Print the indep_X
indep_X
```

```python
# Print the dep_Y
dep_Y
```

```python
# Import classification libraries from feature_importance_classification py file
from feature_importance_classification import *
```

```python
# get best features
ensemble_feature_list=permutation_feature_importance(indep_X,dep_Y,5, feature_names= df2.columns)  
```

```python
ensemble_feature_list
```

```python
# Create empty list to store accuracy for each algorithms
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]
```

```python
# Get all the required details from each algorithm and assign the accuracy to the respective list
# Split the train and test data
X_train, X_test, y_train, y_test=split_scalar(indep_X[ensemble_feature_list],dep_Y) 

classifier,Accuracy,report,X_test,y_test,cm=logistic(X_train,y_train,X_test, y_test)
acclog.append(Accuracy)

classifier,Accuracy,report,X_test,y_test,cm=svm_linear(X_train,y_train,X_test,y_test)  
accsvml.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=svm_NL(X_train,y_train,X_test,y_test)  
accsvmnl.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=knn(X_train,y_train,X_test,y_test)  
accknn.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Navie(X_train,y_train,X_test,y_test)  
accnav.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Decision(X_train,y_train,X_test,y_test)  
accdes.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=random_forest(X_train,y_train,X_test,y_test)  
accrf.append(Accuracy)
```

```python
# get all the accuracy of each algorithm and print it in a dataframe
result=fi_classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)
```
